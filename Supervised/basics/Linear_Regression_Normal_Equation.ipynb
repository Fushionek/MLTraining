```python
import numpy as np
```

```python
X1 = np.array([1, 2, 3, 4, 5, 6])
Y = np.array([3000, 3250, 3500, 3750, 4000, 4250])
m = len(X1)

print(f'Lata pracy: {X1}')
print(f'Wynagrodzenie: {Y}')
print(f'Liczba próbek: {m}')
```

```python
X1 = X1.reshape(m, 1)
print(X1)
print(X1.shape)
```

```python

bias = np.ones((m, 1))
print(bias)
print(bias.shape)
```

```python
X = np.append(bias, X1, axis=1)
print(X)
print(X.shape)
```

```python
np.dot(X.T, X) #liczenie rownania normalnego
```

```python
L = np.linalg.inv(np.dot(X.T, X))
```

```python
L
```

```python
P = np.dot(X.T, Y)
```

```python
P
```

```python
np.dot(L, P)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
regression.fit(X1, Y) #przekazujemy tylko cechy

print(regression.intercept_)
#wspolczynnik w0 miesjce przeciecia z osia Y
#wspolczynnik R2
print(regression.coef_[0])
```
