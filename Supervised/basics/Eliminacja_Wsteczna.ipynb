```python
import numpy as np
from sklearn.datasets import make_regression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

X, y = make_regression(n_samples=100, n_features=10, n_informative=3, noise=0.1, random_state=42)

model = LinearRegression()

selector = RFE(model, n_features_to_select=3, step=1)
selector = selector.fit(X, y)

print("Czy cecha została wybrana? (True/False):")
print(selector.support_)

print("\nRanking cech (1 oznacza wybraną):")
print(selector.ranking_)
```
